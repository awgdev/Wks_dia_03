```python
#Matemática: Valor_maximo = 100, Valor_minimo = 0, Valor_médio = 66.
#Leitura: Valor_maximo = 100, Valor_minimo = 17, Valor_médio = 69,1.
#Escrita: Valor_maximo = 100, Valor_minimo = 10, Valor_médio = 68.
df. describe ()
```

```python
# criando um novo dataframe = df2
# .groupby agrupa a coluna ('race/ethnicity') e exibiu a média para cada grupo em [['math score']] usando a função .mean()
df2 = df.groupby('race/ethnicity')[['math score']].mean()
```

```python
#função df2 exibindo gráfico.
df2.plot (kind = 'bar', subplots = True)
plt.title ('Média de Matemática')
```

```python
df3 = df.groupby('test preparation course')[['reading score']].mean()
```

```python
df3.plot (kind = 'bar', subplots = True)
plt.title('Média de Leitura')
```
